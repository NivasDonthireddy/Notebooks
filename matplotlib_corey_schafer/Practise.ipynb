```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
x  = np.linspace(0,2*math.pi,20)
```

```python
plt.tight_layout()
plt.plot(x,np.sin(x), label='sin(x)')
plt.plot(x,np.cos(x), label='cos(x)')
plt.legend()
plt.grid(True)
```

```python
plt.bar(x,np.sin(x))
plt.bar(x,np.cos(x))
```
