```python
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

#Creating a spiral plot
theta = numpy.linspace(-12,12,20)
X=numpy.sin(theta)
Y=numpy.cos(theta)
Z = numpy.linspace(-2,2,20)

ax.plot(X,Y,Z)
```

```python
import numpy
import random
import matplotlib.pyplot as plt

random.seed(1)
from mpl_toolkits.mplot3d import axes3d

x = [random.normalvariate(1,1) for _ in range(400)]
y = [random.normalvariate(1,1) for _ in range(400)]
z = [random.normalvariate(1,1) for _ in range(400)]

fig = plt.figure()
#figsize argument (takes tuple as input) can be passed into .figure() method to increase the size of the figure
ax = fig.add_subplot(projection='3d')

ax.scatter(x,y,z)

ax.set(xlabel='X-axis',ylabel='Y-axis',zlabel='Z-axis')

#ax.set_xlabel('X-axis')
#Similarly for individual elements

```

`Adding Labels to the plots `

```python
import numpy
import random
import matplotlib.pyplot as plt

random.seed(1)
from mpl_toolkits.mplot3d import axes3d

x = [random.normalvariate(1,1) for _ in range(400)]
y = [random.normalvariate(1,1) for _ in range(400)]
z = [random.normalvariate(1,1) for _ in range(400)]

fig = plt.figure()
#figsize argument (takes tuple as input) can be passed into .figure() method to increase the size of the figure
ax = fig.add_subplot(projection='3d')

ax.scatter(x,y,z, label='First Plot')

ax.set(xlabel='X-axis',ylabel='Y-axis',zlabel='Z-axis')

for data in (x,y,z):
    random.shuffle(data)

    
#Plotting shuffled data
ax.scatter(x,y,z, label='Second Plot')
ax.set(xlabel='X-axis',ylabel='Y-axis',zlabel='Z-axis')

ax.legend(loc='upper left')
#ax.set_xlabel('X-axis')
#Similarly for individual elements

```
