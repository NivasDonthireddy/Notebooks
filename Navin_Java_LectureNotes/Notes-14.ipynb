`Set`

```java
import java.util.HashSet;

Set<Integer> values = new HashSet<>();
values.add(92);
values.add(5);
values.add(6);
System.out.println(values.add(9));
System.out.println(values.add(9));

for(int i:values){
    System.out.print(i+" ");
}
```

The main differnce between Set & List is that , it will contain only unique elements..That's why we are getting false when adding the duplicate entry. This feature only possible with set. 

Also, If you observe the elements you are fetching are not in Sequence. This is because in HashSet it uses a concept of Hashing , what it does is it will use some algorithm using which the nearest value will come first..

..HashSet wont gives values in Sequence...

--TreeSet   But what if you want to use Ascending/Descending Order you can use TreeSet..

```java
import java.util.HashSet;

Set<Integer> values = new TreeSet<>();
values.add(92);
values.add(5);
values.add(6);
System.out.println(values.add(9));
System.out.println(values.add(9));

for(int i:values){
    System.out.print(i+" ");
}
```

We Humans are not comfortable with values right..so we use key and value pairs to identify. To acieve that we use Maps..Map is an Interface it takes two generic values <K, V>.

```java
import java.util.HashMap;
import java.util.Map;

Map<String,String> map = new HashMap<>();
map.put("myName","Nivas");
map.put("sportsperson","Kane");
map.put("actor","VJS");
map.put("food","Idly");

System.out.println(map);
```

Sequence is getting altered right.It is true as HashMaps also use Hashcode which doesnt give values in sequence...

`How to use Enhanced for loop in Map`

```java
System.out.println(map.get("actor"));
System.out.println(map.get("actor1"));
```

```java
Set<String> keys = map.keySet();
// when you have 2 values with same key.. it will replace as its set
// as set takes only unique keys.

for(String key:keys){
    System.out.println(key+": "+map.get(key));
}
```

```java
Map<String,String> phonebook = new HashMap();
phonebook.put("Nivas","7252952909");
phonebook.put("Chintu","9849584717");
phonebook.put("Suresh","7202452904");
phonebook.put("Navin","7245896120");
phonebook.put("Bhargav","8897564906");

System.out.println(phonebook.get("Navin"));
```

`To get all the keys`

```java
phonebook.keySet();
```

`To print all the elements of Map : 2 ways possible`

`1. using Enhanced for loop -- `

```java
Set<String> keys = phonebook.keySet();

for(String key:keys){
    System.out.println(key+" : "+phonebook.get(key));
}
```

`2. Using Map.Entry -- ` Entry is the Interface in Map(Map is an Interface and inside that we have an Interface)(So its a Nested Interface). In Map we have a method called as `.entrySet()`, which will give me set of entries.(key-value pair is one entry). we will be getting set of entries and offcourse we will get entire HashTable right. now we can iterate the set and when we fetch one one value instead of getting single value we get entries that consists of both key and value pair..

```java
Set<Map.Entry<String,String>> values = phonebook.entrySet();
for(Map.Entry<String, String> e:values){
    System.out.println(e.getKey()+": "+ e.getValue());
    e.setValue("III");
}
```
