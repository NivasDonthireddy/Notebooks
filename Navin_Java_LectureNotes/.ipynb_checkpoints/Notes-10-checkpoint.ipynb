`Exception  Handling`

In Java we have `Errors` & `Exceptions`- 

`Different Exceptions`
1. Opening file but file doesnt exist
2. Asking a user for input and user is not giving input.
3. Trying to run a process but CPU is not available
4. Trying to divide a number by zero.

`Notes`<br>

1. Main Class is `Throwable` (generally all 'ables' are interfaces but here its a class)

2. `Exception` & `Error` are the subclasses of Main Class `Throwable`. (Exceptions can be handles as a programmer. but errors we cant handle)<br>
Eg. a. Jvm not working properly , we cant do anything with that.<br>
b. RAM going out of memory.

3. `Exceptions` -- > 2 types `checked` & `unchecked`. When we say its checked we are allowing the user to handle it. If user cannot handle the exception make it unchecked.

4. In java `statements` can be classified --> `normal` & `critical`<br>
int i=1 is a normal statement. but<br>
int k=i/j is a critical statement.

5. Whenever we have critical statements we need to put that statements inside a block and if the block generates any exception we need to handle it.
6. We use try & catch blocks to handle exception.If there is any problem with try block code will go to catch.
7. `Types of Exceptions` - Arithmetic exception, IOexception etc. catch block defined to take one type of exception will only catch that kind of exceptions , therefore we need to define another catch block with IOException as well to catch if the exception if IO type.
8. Now if we dont know the kind of Exception a critical statement throws , we can go for general exception i.e `Exception` which is a superclass of all kinds of exceptions.

9. We can also handle exception using `throws`. i.e sending the request above the stack. we can use `throw` keyword.

I want to print BYE Irrespestive of whether i am able to get answer or not. We can copy the same line and paste it in catch as well.. But that's a redundant line. Redundant code is criminal act.

we can use `finally{}` block to do that which will be called after one of try and catch gets called. 

You should always close the resources you are opening.

```java
import java.io.InputStreamReader;
import java.io.InputStreamReader;

int i,j=1,k;
BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

i=8;
try{
    System.out.println("Enter a number: ");
    j=Integer.parseInt(br.readLine());
    k=i/j;
    System.out.println("Output is : "+k);    

}
catch(IOException e){
    System.out.println("Some IO Error");
}
catch(ArithmeticException e){
    System.out.println("Division by zero");
}

catch(Exception e){
    System.out.println("Unknown Exception");
}

finally{
    br.close();
    System.out.println("Bye");
}
```

`Try with Resources` - Introduced in java 1.7 and later.

We can use this method with sockets,files,database etc. for any resources we can use try with resources concept.

```java
import java.io.InputStreamReader;
import java.io.InputStreamReader;
try{
    String str ="";
    BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
    str = br.readLine();
}
catch(Exception e){
    System.out.println(e);
}
finally{
    br.close();
}
// br is inputOutput operation so we need to give IO exception.
// Whenever we are working with resources we need to close them.

```

If you are creating a finally block just to close the resource , dont create it. Instead of that create a resource inside try parametre(..), so that the entire  code can be reduced.

```java
import java.io.InputStreamReader;
import java.io.InputStreamReader;
try(BufferedReader br = new BufferedReader(new InputStreamReader(System.in))){
    String str ="";
    str = br.readLine();
}
```

```java
int arr[] = {1,2,3,4,5};

        try {
            for(int n=0;n<=arr.length;n++) {
                System.out.print(arr[n]+" ");
            }
        }
        catch(Exception e) {
            //This block will handle all types of exceptions that occur in try block
            System.out.println("Some error occurred!");
        }
        catch(ArithmeticException e) {
            //This block will only execute if any Arithmetic exception occurs in try block
            System.out.println("Divide by Zero exception occurred!");
        }

        catch(ArrayIndexOutOfBoundsException e) {
            System.out.println();
            //This block will only execute if any ArrayIndexOutOfBoundsException occurs in try block
            System.out.println("Array index out of range exception occurred!");
        }

        finally{
            System.out.println("Good Bye");
        }
```

`Types of Exceptions`<br>
There are two different types of exceptions in Java:

`Checked Exception`

All exceptions other than runtime exceptions are known as Checked Exceptions as the compiler checks them during compilation to see whether the programmer has handled them or not. If these exceptions are not handled/declared in the program, you will get a compilation error. In this case, the programmers are forced to either handle these exceptions or declare to allow their propagation.

 E.g. - SQLException, IOException, etc. are Checked Exceptions

`Unchecked Exception`

Unchecked exceptions are runtime exceptions. These exceptions are not checked at compile-time. So, the compiler does not check whether the programmer has handled them or not. In this case, the programmers are not forced to handle or declare their propagation.

E.g. - ArithmeticException, NullPointerException, ArrayIndexOutOfBoundsException, etc. are Unchecked Exceptions

`throw` keyword to forcefully throw an error if certain criteria is not satisfied.<br>
1. `throw new <exceptionclass>("error message");` . This is possible with any object of type `Throwable`
2. If error message is not specified then it will search for exception object of exceptionclass in catch block.
3. if error message is specified and catch block withe Exception is passed then catch block will get preference.
4. In catch block the 'error msg' passed will be stored in `e` . Therefore we can get the same message that is poopped in 1st case if we use `e` inside catch Exception print statement.
5. If you want to print only the error message , then try using `e.getMessage()` inside print statement. Then only the message passed in `error message` will get printed.
