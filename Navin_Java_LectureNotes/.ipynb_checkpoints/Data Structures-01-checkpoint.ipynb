`Arrays - pros & cons of arrays`

Arrays will always give you contigous memory locations. every element will be stored in a memory locations that are adjacent to each other and to fetch te values we use index numbers.

`Pros`

1. Fetching values is faster by index<br>

`Cons`
1. Non Expandable
2. Not Shrinkable

`Dynamic Arrays` - creating double sized array to utilize the size in future . But this will not be memory efficient.
