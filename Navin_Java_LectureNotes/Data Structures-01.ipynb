`Linked List Implementation in Java`

```java
class Node{
    int data;
    Node next;
}

class LinkedList{
    Node head;
    public void insert(int data){
        Node node = new Node();
        node.data = data;
        node.next = null;
        
        if(head==null){
            head = node;
        }
        
        Node n = head;
    }
    
    public void show(){
        Node x = head;
        while(x.next!=null){
            System.out.println(x.data);
            x = x.next;
        }
    }
}

// inside main
LinkedList list = new LinkedList();
list.show();
```

`How to add element to linkedlist at the start`

```java
public void insertAtStart(int data) {
    Node node = new Node();
    node.data = data;

    if(head==null) {
        head = node;
    }
    else {
        node.next = head;
        head = node;
    }
}
```

`Insert elements at any location`

```java
public void insertAt(int index,int data) {
    Node node = new Node();
    node.data = data;
    Node n = head;
    Node temp=null;
    for(int i=1;i<=index;i++) {
        if(i<index) {
            n = n.next; 
        }
        else {
            temp = n.next;
        }
    }

    n.next = node;
    node.next = temp;

}
```

```java
public void deleteAt(int index) {
    Node n1;
    if(index==0) {
        head = head.next;
    }
    else {
        Node n = head;
        for(int i=0;i<index-1;i++) {
            n = n.next;
        }
        n1 = n.next;
        n.next = n1.next;
        n1 = null;
    }
}
```
