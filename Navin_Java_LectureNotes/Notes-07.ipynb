```java
class A{
    int i;
    
    public void show(){
        System.out.println("show method called");
    }
    
    public A(){
        i = 10;
    }
    
    public A(int k){
        i = k;
    }
    
    public A(int k,float j){
        i = (int)k+(int)j;
    }
}

//inside main
A obj = new A();
obj.i
```

```java
A obj = new A(12,12.20f);
obj.i
```

`Static Keyword` - If used static keyword then we dont require an object to call the method or an instance variable. we can directly call the method with the help of ClassName/


```java
class Abc{
    static int i;
    public static void show(){
        System.out.println("Hi");
    }
}

//inside main

Abc obj1 = new Abc();
obj1.show();

Abc.show();
Abc.i;
```

`Count Number of Objects & Classes created`

`i ` is an instance variable.so it will be different for different objects. when an objet is created new instance of `i` is created and is assigned to that object. Therefore counter doesnt increment whenever i create a new object. to overcome this problem we can make the `i` static and this do the job. 

```java
class A{
    int i;
    public A(){
        i++;
    }
    
    public void Counter(){
        System.out.println(i);
    }
}

A obj1 = new A();
A obj2 = new A();
obj1.Counter();
obj2.Counter();
```

```java
class A{
    static int i=0;
    public A(){
        i++;
    }
    
    public void Counter(){
        System.out.println(i);
    }
}

A obj1 = new A();
obj1.Counter();
A obj2 = new A();
obj2.Counter();
```

`Static Block`

Static block gets executed when a class is loaded into jvm.before even calling the class methods static block gets executed first. 

Static members doesnt belong to any objects they are accesible using classes.

Static block gets called when class is loaded. Also, it is possible to create any number of static blocks. All the blocks will be loaded when class in loaded.Multiple static will load in the sequence provided.

```java
class A{
    static{
        System.out.println("Hi");
    }    
    
    public static void main(String[] args){
        System.out.println("Nivas");
    }
}

A obj1 = new A();
```
